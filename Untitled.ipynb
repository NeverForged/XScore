## Theory: Player Segmentation as Scoring Model
This uses the work done in player segmentation, or at least the theory I'm currently going with, to attempt to correctly score an item.
* Seperate the item
* combine all essays
* break into *sentences*, add answer key, process text.
* run LDA on these sentences
* get essays broken into essays
* run lda transform on the essays
* segment the essays into groups using kmeans
* categorize each segment*

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
import string
from autocorrect import spell
from nltk.stem.porter import *
from IPython.display import clear_output


def xscore_process(lstr):
    # stop words and Punctuation
    stop = stopwords.words('english') + list(string.punctuation)
    stemmer = PorterStemmer()


    #  Science, Math = Spelling not counted.
    text = ' ||| '.join(lstr).lower()
    for pun in list(string.punctuation):
        if pun != '|':
            text = text.replace(pun,'')
    check = list(set(text.split(' ')))
    for i, word in enumerate(check):
        clear_output(wait=True)
        percent = (i)/len(check)
        print('{:.3f}% complete'.format(100*percent))
        A = int(50*percent)
        B = 50 - A
        print('[{}{}] Spellchecked...'.format('/'*A, '-'*B))
        print('[{}] Tokenized'.format('-'*50))
        print('[{}] Stemmed'.format('-'*50))
        if not ('|') in word and word.isalpha():
            correct = spell(word)
            if correct != word:
                text = text.replace(word, correct)

    # Tokenize for processing...
    tokenized = []
    lst = text.split('|||')
    for i, essay in enumerate(lst):
        clear_output(wait=True)
        percent = (i)/len(lst)
        print('100% | {:.3f}% complete'.format(100*percent))
        A = int(50*percent)
        B = 50 - A
        print('[{}] Spellchecked!'.format('/'*50))
        print('[{}{}] Tokenized...'.format('/'*A, '-'*B))
        print('[{}] Stemmed'.format('-'*50))
        tokenized.append([i for i in word_tokenize(essay.lower()) if i not in stop])

    #
    # Stem them to avoid parts of speach, etc.
    stemmed = []
    for i, essay in enumerate(tokenized):
        clear_output(wait=True)
        percent = (i)/len(tokenized)
        print(len(tokenized))
        print('100% | 100% | {:.3f}% complete'.format(100*percent))
        A = int(50*percent)
        B = 50 - A
        print('[{}] Spellchecked!'.format('/'*50))
        print('[{}] Tokenized!'.format('/'*50))
        print('[{}{}] Stemmed...'.format('/'*A, '-'*B))
        stemmed.append(' '.join([stemmer.stem(i) for i in essay if i not in stop]))

    clear_output(wait=True)
    print('100% | 100% | 100%')
    print('[{}] Spellchecked!'.format('/'*50))
    print('[{}] Tokenized!'.format('/'*50))
    print('[{}] Stemmed!'.format('/'*50))
    return stemmed
```

```python
n_components = 300
n_clusters = 30
ngram_range_max = 10
```

```python
from sklearn.decomposition import LatentDirichletAllocation
from sklearn.feature_extraction.text import CountVectorizer

# first, process the things
essays = ' '.join(train['EssayText'])

essays = essays + """
You need to know how much vinegar was used in each container.
You need to know what type of vinegar was used in each container.
You need to know what materials to test.
You need to know what size of materials should be used.
You need to know what surface area of materials should be used.
You need to know how long each sample was rinsed in distilled water.
You need to know what drying method to use.
You need to know what size of container to use.
You need to know what type of container to use.
"""

essays = xscore_process(essays.split('.'))

cv = CountVectorizer(ngram_range=(1,ngram_range_max))
vector = cv.fit_transform(essays)

ldas = LatentDirichletAllocation(n_components=n_components, verbose=1, max_iter=1)
ldas.fit(vector)
```

```python
import numpy as np

# now to transform the actual essays...
training = list(train['EssayText'])

# process...
training = xscore_process(training)

# vectorize..
training_vect = cv.transform(training)

# lda
training_topics = ldas.transform(training_vect)

```

```python
from sklearn.cluster import KMeans

# kmeans, 30 to match winner of dataset I'm using.. will eventually add grid-searching
kmeans = KMeans(n_clusters=n_clusters)
distences = kmeans.fit_transform(training_topics)
# distences = training_topics
```

```python
from sklearn.decomposition import PCA as sklearnPCA
import matplotlib.pyplot as plt
import matplotlib.colors

# visualizing the answers...
pca = sklearnPCA(n_components=2)

tt_norm = (training_topics - np.min(training_topics))/(np.max(training_topics) - np.min(training_topics))

visual = pd.DataFrame(pca.fit_transform(tt_norm))
visual['cat'] = kmeans.predict(training_topics)
visual['score'] = train['Score1']

cmap = plt.cm.rainbow
norm = matplotlib.colors.Normalize(vmin=0, vmax=np.max(visual['cat']))

for cat in set(visual['cat']):
    plt.scatter(visual[visual['cat']==cat][0], visual[visual['cat']==cat][1], color=cmap(norm(cat)), label=str(cat))
plt.legend()
plt.show()


```
